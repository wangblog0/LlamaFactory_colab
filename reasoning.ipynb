```python
from pyngrok import ngrok
import os

# 填入你的 ngrok Authtoken
NGROK_TOKEN = "你的_NGROK_AUTHTOKEN"
ngrok.set_auth_token(NGROK_TOKEN)

# 开启隧道，映射到 vLLM 默认的 8000 端口
public_url = ngrok.connect(8000).public_url
print(f"🎉 你的公网 API 地址是: {public_url}")
```

```python
# 你的 Hugging Face 模型 ID (例如: "username/my-llama-model")
MODEL_ID = "你的HuggingFace模型路径"
# 设置你自定义的 API 密钥
CUSTOM_API_KEY = "my_super_secret_key"

# 使用 nohup 后台运行，避免阻塞 Colab
# --dtype auto 自动选择精度，--gpu-memory-utilization 0.9 占用90%显存
import subprocess

cmd = f"python -m vllm.entrypoints.openai.api_server \
    --model {MODEL_ID} \
    --api-key {CUSTOM_API_KEY} \
    --dtype auto \
    --gpu-memory-utilization 0.9"

# 启动服务器
process = subprocess.Popen(cmd, shell=True)
```
